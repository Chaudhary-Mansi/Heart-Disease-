```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
columns_with_outliers = ["BP","cholestrol","max heart rate","ST depression","flouroscopy coloured"]

# Impute outliers for each column
for column in columns_with_outliers:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    
    # Calculate IQR
    IQR = q3 - q1
    
    # Define the lower and upper bounds
    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    
    # Cap the outliers
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

# Display the DataFrame
print(df.head())
```

```python

feature=["BP","cholestrol","max heart rate","ST depression","flouroscopy coloured"]
x=df[feature]
target=["heart disease"]
y=df[target]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=40)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_default = DecisionTreeClassifier()
dt_default.fit(x_train, y_train)
```

```python
from sklearn import tree
tree.plot_tree(dt_default, feature_names=feature)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
ypred=dt_default.predict(x_test)
```

```python
print(classification_report(y_test,ypred))
```

```python
print(confusion_matrix(y_test,ypred))
```
